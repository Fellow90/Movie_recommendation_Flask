```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
movies.head(112)
```

```python
credits.head()
```

```python
movies = movies.merge(credits,on = "title")
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
def convert(object):
    newList = []
    for i in ast.literal_eval(object):
        newList.append(i["name"])
    return newList
```

```python
movies["genres"]=movies["genres"].apply(convert)
```

```python
movies.head()
```

```python
movies["keywords"] = movies["keywords"].apply(convert)
```

```python
movies.head()
```

```python
def convertcast(object):
    newList = []
    count = 0
    for i in ast.literal_eval(object):
        if count!=3:
            newList.append(i["name"])
            count+=1
        else:
            break
    return newList
```

```python
movies["cast"] = movies["cast"].apply(convertcast)
```

```python
movies.head()
```

```python
def convertcrew(object):
    newList = []
    for i in ast.literal_eval(object):
        if i["job"]=='Director':
            newList.append(i["name"])
            break
    return newList
```

```python
movies['crew'] = movies['crew'].apply(convertcrew)
```

```python
movies['overview'].apply (lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply (lambda x:[i.replace(" ","")for i in x])
movies['keywords'] = movies['keywords'].apply (lambda x:[i.replace(" ","")for i in x])
movies['cast'] = movies['cast'].apply (lambda x:[i.replace(" ","")for i in x])
movies['crew'] = movies['crew'].apply (lambda x:[i.replace(" ","")for i in x])
movies["overview"] = movies["overview"].apply (lambda x:x.split())
```

```python
movies['tags'] = movies["overview"] + movies["genres"] + movies["keywords"] + movies["cast"] + movies["crew"]
```

```python
movies.head()
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
new_df["tags"] = new_df["tags"].apply(lambda x:" ".join(x))
```

```python
new_df['tags'] = new_df['tags'].apply (lambda x:x.lower())
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
       y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words='english',max_features=2000)
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
def recommender(movie):
    idx = new_df.loc[new_df['title'] == movie].index
    if len(idx) == 0:
        print(f"Movie '{movie}' not found!!")
    else:
        idx = idx[0]
        print('Movie Selected: ',new_df['title'][idx], 'Index: ',idx)
        print('Searching for recommendations.....')
        try:
            distances = similarity[idx]
            indices = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[:11]
            for i in indices:
                print(new_df.iloc[i[0]].title)
        except IndexError:
            print('Similar movies are not found in the dataset.')

```
